# SVD Matrix Completion

```python
import numpy as np
from matrix_completion import svt_solve, calc_unobserved_rmse
import matplotlib.pyplot as plt
```

```python
U = np.random.randn(5, 100)
V = np.random.randn(3, 100)
R = np.random.randn(5, 3) + np.dot(U, V.T)

mask = np.round(np.random.rand(5, 3))
R_hat = svt_solve(R, mask)

print("RMSE:", calc_unobserved_rmse(U, V, R_hat, mask))
```

```python
plt.figure()
plt.imshow(R);plt.colorbar()
plt.figure()
plt.imshow(R_hat);plt.colorbar()
```

```python
plt.imshow(R-R_hat);plt.colorbar()
print(np.linalg.norm(R-R_hat))
```
