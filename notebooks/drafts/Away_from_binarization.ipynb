# From Binarization to Nuance


## When Binarization Fails
Let's start with a signal and try to binarize it into one of two hypotheses.

```python
import numpy as np
import scipy.signal as sig

x = np.linspace(-10,10,1000)
y = np.sin(2 * np.pi * 1 * x) + 3/3 * np.sin(2 * np.pi * 3 * x)

import matplotlib.pyplot as plt
plt.figure()
plt.plot(x,y)
plt.xlim((-10,-9))

y1 = np.sin(2 * np.pi * 1 * x)
y2 = 1/3 * np.sin(2 * np.pi * 3 * x)
plt.plot(x,y1,linestyle='dotted')
plt.plot(x,y2,linestyle='dotted')
```

Obviously that's a little silly to do, we should be a bit more nuanced. Instead of asking if our signal looks like either of the templates we can ask if some combination of our templates look like the signal.

> Knowing what we know, this is a good way to go...
